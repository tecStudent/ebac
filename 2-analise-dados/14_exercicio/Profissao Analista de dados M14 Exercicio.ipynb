## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = sns.barplot(x="cut", y="price", data=data)

diamonds.axes.set_title("Relação de corte com preço", fontsize=12, fontweight="bold")
```

**Insight do gráfico 1**: Existe pouca diferença de preço entre os cortes
